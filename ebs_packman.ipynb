```
"""
Utility functions to enable video recording of gym environment and displaying it
To enable video, just do "env = wrap_env(env)""
"""

def show_video():
    mp4list = glob.glob('video/*.mp4')
    if len(mp4list) > 0:
        mp4 = mp4list[0]
        video = io.open(mp4, 'r+b').read()
        encoded = base64.b64encode(video)
        ipythondisplay.display(HTML(data='''<video alt="test" autoplay 
                    loop controls style="height: 400px;">
                    <source src="data:video/mp4;base64,{0}" type="video/mp4" />
                 </video>'''.format(encoded.decode('ascii'))))
    else: 
        print("Could not find video")
    

def wrap_env(env):
    env = Monitor(env, './video', force=True)
    return env
```

```
# load model
model = ACER.load('stable_baselines_acer_cnnlnlstm.pkl', lr_schedule='constant', verbose=1)
```

```
# set up evaluation environment for video recording
env = wrap_env(gym.make('MsPacman-v0'))
env = MaxAndSkipEnv(env, skip=4)
env = wrap_deepmind(env, episode_life=False, clip_rewards=False, frame_stack=True)
env = DummyVecEnv([lambda: env])
```

```
# generate sample video
observations = env.reset()
while True:
    env.render(mode='rgb_array')
    actions, _states = model.predict(observations)
    observations, rewards, dones, infos = env.step(actions)
    if isinstance(dones, np.ndarray):
        dones = dones.any()
    if dones:
        break
env.close()
show_video()
```

```
# set up evaluation environment for statistical analysis over multiple episodes
env = gym.make('MsPacman-v0')
env = MaxAndSkipEnv(env, skip=4)
env = wrap_deepmind(env, episode_life=False, clip_rewards=False, frame_stack=True)
env = DummyVecEnv([lambda: env])
```

```
# run 20000 tests and generate a list of final rewards
number_of_tests = 20000

final_rewards = []
for i in range(number_of_tests):
    observations = env.reset()
    final_reward = 0
    while True:
        actions, _states = model.predict(observations)
        observations, rewards, dones, infos = env.step(actions)
        final_reward += sum(rewards)
        if isinstance(dones, np.ndarray):
            dones = dones.any()
        if dones:
            break
    final_rewards.append(final_reward)
cumulative_final_rewards = [sum(final_rewards[:i + 1]) for i in range(number_of_tests)]
```

```
# final evaluation
plt.rcParams["figure.figsize"] = (10, 10)

plt.subplot(3, 1, 1)
plt.scatter(range(1, number_of_tests + 1, 1), final_rewards, s=0.5)
plt.xlabel('episode')
plt.ylabel('final score')
plt.title('ACER + CnnLnLstm - Total Rewards: {:,}'.format(cumulative_final_rewards[-1]))

plt.subplot(3, 1, 2)
plt.plot(range(1, number_of_tests + 1, 1), cumulative_final_rewards)
plt.xlabel('episode')
plt.ylabel('cumulative final score')

plt.subplot(3, 1, 3)
plt.hist(final_rewards, bins=50)
plt.xlabel('final score')
plt.ylabel('count')

plt.tight_layout()
plt.show()
```
